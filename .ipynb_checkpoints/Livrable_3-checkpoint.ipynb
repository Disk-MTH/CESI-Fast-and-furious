## <font color = darkred> I.La pente

Pour ne pas t'alourdir avec plein de code dans tous les sens surtout que dans cette présentation jupyter, je vais devoir mettre beaucoup trop de code de dépendance qui va polluer ta lecture. Le code ne sera donc pas executable ici, je vais te mettre uniquement les appels de mon code et les références de lignes à vérifier. Je te mettrai l'entièreté du code en annexe ainsi qu'en libre accès sur mon [Github](https://github.com/Disk-MTH/CESI-Fast-and-furious) afin que tu puisses en prendre connaissance(y comprit de la documentation présente sour la forme de docstrings directement dans le code) et l'exécuter en parallèle.

Pour commencer doucement, voici une partie du code qui devrait t'intéresser

```python
# Je commence par définir les paramètres des différentes études afin de pouvoir contrôler ma modélisation facilement

study_duration = 1.5  # in seconds
study_interval = 0.01  # in seconds
car_mass = 0.03  # in kg
friction_coefficient = 0.002  # coefficient of friction with the ground

slope_height = 0.93  # in meters
slope_angle = 40  # in degrees
slope_v0 = (0, 0)  # v0x, v0y in m/s
slope_p0 = (0, 0)  # p0x, p0y in m

slope_ceil = 5  # in meters
slope_height_interval = 0.01  # in meters

looping_radius = 0.115  # in meters
looping_p0 = 90  # in degrees

ravine_length = 0.7  # in meters
ravine_height = 0.1  # in meters
ravine_drag_coefficient = 0.001  # coefficient of drag * surface
ravine_lift_coefficient = 0.001  # coefficient of lift * surface
```

<font color = #006400> Comme lors des précédents rapports, la pente fait office d'échauffement dans nos études. Son utilité est de donner un maximum de vitesse afin de passer les épreuves suivantes : à savoir le looping suivi du ravin.
Pour encore plus de lisibilité dans toutes les études suivantes, on a décidé de tout afficher dans un graphique aux petits oignons.

<font color = #006400> Pour te rafraîchir la mémoire, jette un oeil au schéma de la dernière fois, ça devrait te rappeler des choses pas vrai ?

<font color = #006400> A l'aide de projections on obtient les valeurs suivantes pour les trois forces présentes ( Poids [P], Reaction [N] et Frottements [f] ). Il est important de rappeler que les équations de gauche ne prennent pas en compte les frottements de la situation. On peut également en déduire qu'il n'y a pas d'accélération sur y :



<font color = #006400> On peut appliquer la seconde loi de Newton (aussi nommée PFD) et en déduire l'accélération sur <i> x </i> et <i> y </i> puis intégrer et récupérer la vitesse ainsi que la position sur ces deux composantes :


```python
# Ici, c'est assez simple, je crée un objet de type Slope (c'est une classe de ma conception qui va représenter notre pente et qui va donc être réactive aux changements afin de voir l'influence de tel ou tel paramètre) et je lui passe les paramètres que je veux
slope = Slope(duration=study_duration, interval=study_interval, alpha=slope_angle, v0=slope_v0, p0=slope_p0,
              mu=friction_coefficient, is_alpha_degrees=True)

# J'appelle la fonction servant à dessiner les courbes de la pente (accélération, vitesse et position), avec un argument optionnel qui permet de tracer une droite verticale à un instant "t" ce qui aura pour effet d'afficher les valeurs aux croisement de la droite et des courbes
slope.trace(slope.get_t_at_p(slope_height / np.sin(np.radians(slope_angle))))

# J'appelle la fonction qui va me tracer la courbe de vitesse en fonction de la hauteur, pour que tu puisse te rendre compte de l'impact qu'a la hauteur de la pente sur la vitesse en bas
slope.trace_v_end_terms_of_height(ceil=slope_ceil, interval=slope_height_interval)

# Je stocke la vitesse en bas de la pente et je l'affiche
v_at_slope_end = slope.get_v_end_terms_of_height(slope_height)[2]
print("Velocity at the end of the slope: {} m/s".format(v_at_slope_end))
```

## <font color = darkred> II.Le looping

<font color = #006400> On associe alors les deux équations horaires à une seule équation-différentielle que l'on pourra ensuite résoudre numériquement :
    

```python
# On crée un objet de type Looping (c'est une classe de ma conception qui va représenter notre looping et qui va donc être réactive aux changements afin de voir l'influence de tel ou tel paramètre)
looping = Looping(duration=study_duration, interval=study_interval, radius=looping_radius, v0=v_at_slope_end,
                  p0=looping_p0, mu=friction_coefficient, is_v0_ms=True, is_p0_degrees=True)

# On trace les courbes de l'accélération, de la vitesse et de la position du looping dans différentes unités (je ne suis pas sûr que les radians par secondes soient ta tasse de thé, j'ai donc mis des options pour pouvoir choisir les unités)
looping.trace()
looping.trace(p_in_degrees=True, v_in_ms=True)

# Là, je fais l'affichage de quelques valeurs clé du looping que tu peux obtenir avec mon code
print("Minimum velocity for looping: {} m/s".format(looping.get_min_velocity(v_in_ms=True)))
print("Velocity at the end of the looping: {} m/s".format(looping.get_v_after_looping(v_in_ms=True)))

# Ici, je fais une boucle qui va diminuer la hauteur de la pente jusqu'à ce que la vitesse minimale du looping soit atteinte afin de savoir quelles est la hauteur théorique minimale de la pente pour passer le looping
while looping.get_min_velocity(v_in_ms=True) < v_at_slope_end:
    slope_height -= 0.1
    v_at_slope_end = slope.get_v_end_terms_of_height(slope_height)[2]

print("Minimum height of the slope to pass looping: {} m".format(slope_height))
```
